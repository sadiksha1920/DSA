Change 'Period' to datetime

```python
df.groupby(["Series_title_1", "STATUS"])["value"].agg(
    ["mean", "min", "max", "std"]
)
```

Numpy Normalization

```python
values = df["value"].to_numpy() #convert pandas series => numpy array.
```

```python
mean_val = np.mean(values)
std_val= np.std(values)
```

```python
df["value_normalized"] = (values - mean_val) / std_val
```

Standardization (Z-score).

Required for many ML models.

Data Visualization

Missing Values Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
sns.heatmap(df.isna(),cbar=False)
plt.title("Missing value Heatmap")
plt.show()
```

Distribution

```python
#histogram
import matplotlib.pyplot as plt

plt.hist(df["value"], bins=100)
plt.title("Distribution of Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Range of data

Concentration zones

Rare values

```python
#CDF (Cumulative Distribution)

sorted_vals = np.sort(df["value"])
cdf = np.arange(len(sorted_vals)) / len(sorted_vals)

plt.plot(sorted_vals, cdf)
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Values")
plt.show()
```
