```python
import numpy as np

#input
x = np.array([2.0])

#parameters
w = np.array([0.5])
b=1.0
#forward propagation
z = np.dot(w,x)+b
y_hat = z #no activation(linear)

print("Z:", z)
print("prediction:",y_hat)
```

```python
#with activation

def relu(z):
  return np.maximum(0,z)

z = np.dot(w,x) +b
a= relu(z)

print("Activated output :",a)
```

```python
#multilayer

#inputs
X = np.array([[1.0,2.0]])

#weights and biases
W1 = np.array([[0.1,0.2],[0.3,0.4]])
b1 = np.array([0.1,0.1])
W2 = np.array([[0.5],[0.6]])
b2 = np.array([0.2])

#layer1
z1 = np.dot(X,W1)+b1
a1 = np.maximum(0,z1) #ReLU

#layer 2
z2 = np.dot(a1,W2)+b2
y_hat = z2

print("Prediction:",y_hat)

```

```python
#loss function\\
y = np.array([[10.0]])

loss = np.mean((y-y_hat)**2)
print("Loss:",loss)
```

```python
import numpy as np

# input, target
x = np.array([1.0, 2.0])     # 2 input features
y = 4.0                      # true output

# weights and bias
w = np.array([0.1, 0.2])     # 2 weights
b = 0.0

# learning rate
lr = 0.01

for epoch in range(10):
    # forward
    y_hat = np.dot(w, x) + b
    loss = (y - y_hat) ** 2

    # backward
    dL_dyhat = -2 * (y - y_hat)
    dL_dw = dL_dyhat * x
    dL_db = dL_dyhat

    # update
    w -= lr * dL_dw
    b -= lr * dL_db

    print(f"Epoch {epoch+1}, Loss: {loss:.4f}, y_hat: {y_hat:.4f}")
```

```python
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
#Load and prepare dataset
data  = load_iris()
X= data.data
y=  data.target.reshape(-1,1)
```

```python
# One-hot encode labels
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y)
```

```python
#Split Dataset
X_train, X_test,y_train,y_test = train_test_split(X,y_onehot,test_size=0.2,random_state=42)
```

```python
#convert to tensors
X_train = tf.constant(X_train,dtype=tf.float32)
y_train = tf.constant(y_train,dtype=tf.float32)
X_test = tf.constant(X_test,dtype=tf.float32)
y_test = tf.constant(y_test,dtype=tf.float32)
```

```python
#Initialize weights and biases

input_size = X_train.shape[1]
hidden_size = 5
output_size = y_train.shape[1]
```

```python
#random initialization
W1 = tf.Variable(tf.random.normal([input_size,hidden_size]))
b1 = tf.Variable(tf.zeros([hidden_size]))
W2 = tf.Variable(tf.random.normal([hidden_size,output_size]))
b2 = tf.Variable(tf.zeros([output_size]))
```

```python
#learning rate
lr =0.1
epochs =1000
```

```python
import tensorflow as tf

# Training loop
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Forward propagation
        Z1 = tf.matmul(X_train, W1) + b1
        A1 = tf.sigmoid(Z1)

        Z2 = tf.matmul(A1, W2) + b2
        A2 = tf.sigmoid(Z2)

        # Compute loss (Mean Squared Error)
        loss = tf.reduce_mean(tf.square(y_train - A2))

    # Backward propagation (automatic gradient calculation)
    gradients = tape.gradient(loss, [W1, b1, W2, b2])

    # Update weights
    W1.assign_sub(lr * gradients[0])
    b1.assign_sub(lr * gradients[1])
    W2.assign_sub(lr * gradients[2])
    b2.assign_sub(lr * gradients[3])

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy():.4f}")

# Prediction function
def predict(X):
    A1 = tf.sigmoid(tf.matmul(X, W1) + b1)
    A2 = tf.sigmoid(tf.matmul(A1, W2) + b2)
    return tf.argmax(A2, axis=1)

# Predictions
y_pred = predict(X_test)
y_true = tf.argmax(y_test, axis=1)

# Accuracy
accuracy = tf.reduce_mean(tf.cast(y_pred == y_true, tf.float32))
print(f"Test Accuracy: {accuracy.numpy() * 100:.2f}%")
```
