```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
words = ["Ability","shifted","raining","adjustable","reading","risen","rose","ate"]

for word in words:
  print(word,"|", stemmer.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Abiility shifted raining adjustble reading risen rose ate")

for token in doc:
  print(token, "|",token.lemma_)
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text ="the students are Studying Natural Language Processing"

#Tokenize
tokens = word_tokenize(text)

#losd stop words
stop_words = set(stopwords.words('english'))

#remove stop words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Original tokens:",tokens)
print("After stop-word removal:",filtered_tokens)
```

```python
custom_stopwords = {"the","is","are","in","to"}

filtered = [w for w in tokens if w.lower() not in custom_stopwords]
print(filtered)
```
