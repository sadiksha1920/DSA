```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# sample data
X = np.array([
    [2.5,2.4],
    [0.5,0.7],
    [2.2,2.9],
    [1.9,2.2],
    [3.1,3.0],
    [2.3,2.7],
    [2.0,1.6],
    [1.0,1.1],
    [1.5,1.6],
    [1.1,0.9]
])

#Standarize
X_scaled = StandardScaler().fit_transform(X)

#PCA
pca= PCA(n_components=1)
X_pca =pca.fit_transform(X_scaled)

print("Reduced data shape:",X_pca.shape)
print("Explained variance", pca.explained_variance_ratio_)
```

```python
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

plt.scatter(X_pca_2d[:,0],X_pca_2d[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.show()
```

```python
#chge different color

pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

colors = np.arange(len(X_pca_2d))   # different value for each point

plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=colors, cmap='rainbow')
plt.xlabel("PC1",color='blue')
plt.ylabel("PC2",color='green')
plt.title("PCA Projection")
plt.show()
```
