```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data = pd.DataFrame({
    "study_hours": [2,3,4,5,6,7,1,8],
    "attendance": [60,65,70,80,85,90,50,95],
    "practice_tests": [0,1,1,2,3,3,0,4],
    "pass_exam": [0,0,0,1,1,1,0,1]
})
```

```python
X = data[["study_hours", "attendance", "practice_tests"]]
y = data["pass_exam"]
```

```python
#modle
rf = RandomForestClassifier(
    n_estimators =100,
    max_depth=4,
    random_state=42

)
rf.fit(X,y)
```

Prediction

```python
new_students = pd.DataFrame({
    "study_hours": [3,6],
    "attendance": [95,80],
    "practice_tests": [1,3]
})

rf.predict(new_students)
```

Feature importance

How often and how usefully a feature was used across trees.

```python
for name, importance in zip(X.columns, rf.feature_importances_):
  print(name,round(importance,2))
```

Visualization

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

data = pd.DataFrame({
    "study_hours": [2,3,4,5,6,7,1,8],
    "attendance": [60,65,70,80,85,90,50,95],
    "practice_tests": [0,1,1,2,3,3,0,4],
    "pass_exam": [0,0,0,1,1,1,0,1]
})

X = data[["study_hours", "attendance", "practice_tests"]]
y = data["pass_exam"]

rf= RandomForestClassifier(
    n_estimators = 10,           #keep things small for visualization
    max_depth= 3,
    random_state=42
)
rf.fit(X,y)
```

One tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# pick the first tree
tree = rf.estimators_[0]

plt.figure(figsize=(14,6))
plot_tree(
    tree,
    feature_names=X.columns,
    class_names=["fail","Pass"],
    filled=True,rounded=True
)
plt.show()
```

Multiple Trees

```python
plt.figure(figsize=(20,10))

for i in range(3):
  plt.subplot(1,3,i+1)
  plot_tree(
      rf.estimators_[i],
      feature_names=X.columns,
      class_names=["Fail","Pass"],
      filled =True
  )
  plt.title(f"Tree{i+1}")
  plt.show()
```

Feature importance

```python
import matplotlib.pyplot as plt

importances =rf.feature_importances_

plt.bar(X.columns,importances)
plt.xlabel("Features")
plt.ylabel("Imporance")
plt.title("Random Forest Feature Importance")
plt.show()
```

How trees vote

```python
import numpy as np

new_student = pd.DataFrame({
    "study_hours": [6],
    "attendance": [75],
    "practice_tests": [2]

})

votes = []

for i, tree in enumerate(rf.estimators_):
  pred = tree.predict(new_student)[0]
  votes.append(pred)
  print(f"Tree{i+1}->","pass" if np.mean(votes)>0.5 else "Fail")
```

feature importance

how often and how usefully a feature was used across trees

```python
for name, importance in zip(X.columns,rf.feature_importances_):
  print(name,round(importance,2))
```
