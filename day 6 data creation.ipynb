```python
import pandas as pd

data={
    "Hours":[1,2,3,4,5,6,7,8],
    "Score":[20,30,35,45,55,65,75,85],
    "Pass":[0,0,0,0,1,1,1,1]
}
df= pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours"],df["Score"], c=df["Pass"])
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Pass/Fail Distribution")
plt.show()
```

Representing Students

Each student is a feature vector

Example: Hours=4,Score=55 gives x=(4,55)

svm treats every student as a point in 2D space

Decision Boundary

-> SVM draws a line separating classes

->Equation: w1(hours)+ w2(score) +b=0

-> Predicts Pass(1) or fail(0) based on side of line






Making Prediction

```python
from sklearn.svm import SVC

X = df[["Hours","Score"]]
y = df["Pass"]

svm_linear = SVC(kernel="linear")
svm_linear.fit(X,y)

#Predict new student

new_student = [[4,55]]
prediction = svm_linear.predict(new_student)
print("prediction(0=F,Predictionail,1=pass):",prediction)
```
